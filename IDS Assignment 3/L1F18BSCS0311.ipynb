```python
each_genres_total_money_making = new_movie_raw_data.groupby('Genre (1)',as_index=False)[['Budget ($)','Box Office Revenue ($)']].agg('sum')
```

```python
each_genres_total_money_making
```

```python
plt.plot(each_genres_total_money_making['Genre (1)'], each_genres_total_money_making['Budget ($)'], label = "Budget $")
plt.plot(each_genres_total_money_making['Genre (1)'], each_genres_total_money_making['Box Office Revenue ($)'], label = "Box Office Revenue")
plt.legend()
plt.rcParams["figure.figsize"] = (10,10)
plt.show()
```
